```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset("tips")
iris = sns.load_dataset("iris")
titanic = sns.load_dataset("titanic")
planets = sns.load_dataset("planets")
```

```python
tips
```

## Scatterplot

```python
sns.scatterplot(x="tip",y="total_bill", data= tips)
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day")
```

```python
sns.scatterplot(x="tip", y="total_bill", data = tips, hue="day", size="size")
```

```python
sns.scatterplot(x="tip",y="total_bill",data=tips,hue="day",size="size",palette="YlGnBu")
```

## Histograms

```python
tips
```

```python
sns.histplot(tips["tip"], bins=20)
```

```python
sns.histplot(tips["tip"], kde=True, bins=20) #kde for line
```

```python
sns.histplot(tips["total_bill"], kde = True, bins=20)
```

## distplots

```python
sns.distplot(tips["total_bill"], kde=True, bins=20)
```

## displots

```python
sns.displot(tips["total_bill"],kde=True,bins=20)
```

## Barplots

```python
sns.barplot(x="sex", y="tip", data=tips, palette="YlGnBu")
```

##Boxplot

```python
sns.boxplot(x="day", y ="tip", data=tips, hue ="sex", palette="YlGnBu")
```

```python
sns.boxplot(x="day", y ="total_bill", data=tips, hue ="sex", palette="YlGnBu")
```

## Stripplot

```python
sns.stripplot(x="day", y="tip", data = tips, hue="sex") #sex will not get separated because dodge is not set True
```

```python
sns.stripplot(x="day", y="tip", data = tips, hue="sex", dodge= True)
```

## Jointplots

```python
sns.jointplot(x="tip", y="total_bill",data=tips, kind="reg") #reg for regression
```

```python
sns.jointplot(x="tip", y ="total_bill", kind="kde", data=tips)
```

```python
sns.jointplot(x="tip", y ="total_bill", kind="kde", data=tips, shade=True)
```

```python
sns.jointplot(x="tip", y ="total_bill", kind="kde", data=tips, shade=True, cmap="YlGnBu")
```

```python
sns.jointplot(x="tip", y ="total_bill", kind="hex", data=tips, cmap="YlGnBu")
```

##Pairplot

##Heatmap

```python
titanic.corr() #we want to look at the correlations among the correlations
```

```python
sns.heatmap(titanic.corr(), annot = True, cmap= "YlGnBu") #annot means annotations, values of corr will be printed on the map
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap="coolwarm" ) #colors are relatable
```

```python
sns.heatmap(titanic.corr(), annot=True, cmap="icefire" ) #colors are relatable
```
